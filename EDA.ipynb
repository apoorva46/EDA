```python
# list of numerical variables
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O'] #!= object 

print('Number of numerical variables: ', len(numerical_features))

# visualise the numerical variables
dataset[numerical_features].head()
```

```python
# list of variables that contain year information

year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
year_feature
```

```python
## We will check whether there is a relation between year the house is sold ie. YrSold and the sales price

dataset.groupby('date')['amount'].median().plot() 
# We r grouping by YrSold feature and then
# we r actually considering w.r.t. each and every group for this YrSold,
#We try to find out the median of the SalePrice and then we plot it.
plt.xlabel('date')
plt.ylabel('amount')
plt.title("date vs amount")

```

```python
# Discrete Variable for this Dataset:

discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and
                  feature not in year_feature+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
# Let's see all the discrete features 
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
## Lets Find the realtionship between all the 17 descrete Features and Sale PRice

for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['amount'].median().plot.bar()  # Here grouping by feature of all the decrete features and then
                                                        # i'm comparing with the SalesPrice median.
    plt.xlabel(feature)
    plt.ylabel('amount')
    plt.title(feature)
    plt.show()
```

```python
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
print("Continuous feature Count {}".format(len(continuous_feature)))
```

```python
## Again we should check relationship b/w continuous variables and Dependent feature:
## Since these are continuous variables therefore we will talk about distribution (histogram) here
## Lets analyse the continuous values by creating histograms to understand the distribution

for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature]) # we do the log transformation 
        # We can plot the boxplot using seaborn or boxplot also present in dataframe and we can draw it as:
        
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
        
```
